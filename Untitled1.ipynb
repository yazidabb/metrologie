```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define a function to load and process data from a file
def load_and_process_data(file_path, initial_size=200):
  columns = ['Traverse', 'Force', 'Temps']
  data = pd.read_csv(file_path, skiprows=8, sep='\t', decimal=',', names=columns)
  data = data[['Traverse', 'Force']].dropna()
  data['Traverse'] = pd.to_numeric(data['Traverse'], errors='coerce')
  data['Force'] = pd.to_numeric(data['Force'], errors='coerce')
  data = data.dropna()
  data['Deformation_Percentage'] = (data['Traverse'] / initial_size) * 100
  data['tnacit']=(data['Force']*100)/427.24
  return data

# File paths
file_paths = ["/content/DAQ-Traverse;  - (A intervalle de temps fixe)1.txt","/content/DAQ-Traverse;  - (A intervalle de temps fixe)2.txt","/content/DAQ-Traverse;  - (A intervalle de temps fixe)3.txt" ]
colors = ['b', 'r', 'g']  # Colors for each plot
labels = ['Essai 1', 'Essai 2', 'Essai 3']  # Labels for each plot


# Create the plot
plt.figure(figsize=(10, 6))

# Loop through files and plot data
for file_path, color, label in zip(file_paths, colors, labels):
  data = load_and_process_data(file_path)
  plt.plot(data['Deformation_Percentage'], data['tnacit'], label=label, color=color, linewidth=2)

# Customize the plot
plt.title("Courbe de tnacit en fonction de la Dformation", fontsize=14)
plt.xlabel("Dformation (%)", fontsize=12)
plt.ylabel("Tnacit (CN/Tex)", fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()

# Function to calculate statistics and create table
def create_summary_table(data_list, labels):
    results = []
    for data, label in zip(data_list, labels):
        # Calculate statistics
        max_deformation = data['Deformation_Percentage'].max()
        rupture_deformation = data['Deformation_Percentage'].iloc[-1]
        max_force = data['Force'].max()
        rupture_force = data['Force'].iloc[-1]
        max_Tnacit= data['tnacit'].max()
        # Assuming modulus calculation (replace with your actual calculation)
        modulus = (max_Tnacit/ max_deformation) * 100

        results.append([label, max_deformation, rupture_deformation, max_force,
                        rupture_force, max_Tnacit, modulus])

    # Create dataframe for the table
    columns = ['Essai', 'Dformation Max (%)', 'Dformation Rupture (%)',
               'Force Max (N)', 'Force Rupture (N)', 'Tnacit max (MPa)',
               'Module (MPa)']
    results_df = pd.DataFrame(results, columns=columns)

    # Calculate statistics
    mean_values = results_df.mean(numeric_only=True)
    std_values = results_df.std(numeric_only=True)
    cv_values = (std_values / mean_values) * 100  # CV%
    # LPE and P% at 95% (using t-distribution)
    n = len(results_df)
    t_value = stats.t.ppf(0.975, n - 1)  # 95% confidence
    lpe_values = t_value * (std_values / np.sqrt(n))
    p95_values = mean_values + lpe_values

    # Add statistics to the dataframe
    results_df.loc['Moyenne'] = mean_values
    results_df.loc['Ecart Type'] = std_values
    results_df.loc['CV%'] = cv_values
    results_df.loc['LPE'] = lpe_values
    results_df.loc['P%  95%'] = p95_values

    return results_df

# Separate data for chaine and trame
chaine_data = [load_and_process_data(file_path) for file_path in file_paths]


# Create tables
chaine_table = create_summary_table(chaine_data, labels)


# Display tables
display(chaine_table)
```
